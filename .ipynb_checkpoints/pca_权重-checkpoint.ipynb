```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
traindata_0 = pd.read_csv('cym_train_titanic.csv',index_col='id')
traindata_0.info()
traindata_0.head()
# 对缺失数据的声明
# 先使用常规方法处理缺失数据。本期重点在PCA
```

```python
traindata = traindata_0.copy()
# replace
traindata_0.diagnosis = traindata.diagnosis.replace({'B':0,'M':1}) # one hot
# drop
traindata.drop(columns=['diagnosis'],inplace=True)
traindata.info()
print(traindata.head())

traindata.to_csv('cym_data_pca.csv') 
```

```python
data=pd.read_csv('cym_data_pca.csv',index_col='id')
# data=[[-10,2,66,-1], [-2,6,58,-1], [-3,8,45,-2], [1,9,36,1], [2,10,62,1], [3,5,83,2]]
print(type(data))
data.describe()
```

```python
# normalization
def norm_(x):
    xmean = np.mean(x,0)
    std = np.std(x,0)
    var =np.var(x,0)
    w=var/np.var(x,0).sum()
    print(w.sum())
    return ((x-xmean)*w)/std
data_ = norm_(data)
data_.describe()
```

```python
# V
# ew, ev = np.linalg.eig(data_.T.dot(data_))
ew, ev = np.linalg.eig(np.cov(data_.T))
ew_order = np.argsort(ew)[::-1]
ew_sort = ew[ew_order]
ev_sort = ev[:,ew_order]
# print(ew_sort)
# print(ev_sort)
pd.DataFrame(ew_sort).plot(kind='bar')
print("增加权重,贡献率",ew_sort/ew_sort.sum())
pd.DataFrame(ew_sort/ew_sort.sum()).plot(kind='bar')
# V
V = ev_sort[:,:2]
# Xnew
X_new = data_.dot(V)
# print("x_new",X_new)
X_new.shape
```

```python
# scatter
get_ipython().run_line_magic('matplotlib', 'notebook')
sc = plt.scatter(X_new.iloc[:,0],X_new.iloc[:,1],s=5,c=traindata_0.diagnosis,cmap=plt.cm.coolwarm)
plt.xlabel('PC 0')
plt.ylabel('PC 1')
plt.colorbar(sc)
```
