```python
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
```

```python
#读取数据
X=pd.read_csv('cym_data_pca.csv',index_col='id')
# X = np.array([[-10,2,66,-1], [-2,6,58,-1], [-3,8,45,-2], [1,9,36,1], [2,10,62,1], [3,5,83,2]])  #导入数据，维度为4
# X = np.array([[10000,1,1,1], [10000,3,4,5],[10000,20,200,2000]])  #导入数据，维度为4
print(X)
```

```python
#pca降维
pca = PCA(n_components=4)   #降到2维
pca.fit(X)                  #训练
newX=pca.fit_transform(X)   #降维后的数据
# PCA(copy=True, n_components=2, whiten=False)
print("原始pca,输出贡献率",pca.explained_variance_ratio_)  #输出贡献率
print("降维后数据",newX)                  #输出降维后的数据
print("具有最大方差的成分",pca.components_)
print("所保留的n个成分各自的方差",pca.explained_variance_)
# print("所保留的n个成分各自的方差百分比",pca.explained_variance_ratio_)
# print("所保留的成分个数n",pca.n_components_)
print("噪声方差大小",pca.noise_variance_)
# print("特征均值",pca.mean_)
# print("转换为原始数据",pca.inverse_transform(newX))
# print("转换为降维数据",pca.transform(X))
```

```python
#增量PCA(IPCA)
#使用np.array_split（）方法的IPCA
from sklearn.decomposition import IncrementalPCA
#读取数据
X=pd.read_csv('cym_data_pca.csv',index_col='id')
# print(X)
n_batches = 100
inc_pca = IncrementalPCA(n_components=4)
for X_batch in np.array_split(X, n_batches):
    inc_pca.partial_fit(X_batch)
X_mnist_reduced = inc_pca.transform(X)
print("输出贡献率",inc_pca.explained_variance_ratio_)  #输出贡献率
print("降维后数据",X_mnist_reduced)                  #输出降维后的数据
print("具有最大方差的成分",inc_pca.components_)
print("所保留的n个成分各自的方差",inc_pca.explained_variance_)
print("噪声方差大小",inc_pca.noise_variance_)
```

```python
#随机PCA
#读取数据
X=pd.read_csv('cym_data_pca.csv',index_col='id')
# print(X)
rnd_pca = PCA(n_components=4, svd_solver="randomized")
X_reduced = rnd_pca.fit_transform(X)
print("输出贡献率",rnd_pca.explained_variance_ratio_)  #输出贡献率
print("降维后数据",X_reduced)                  #输出降维后的数据
print("具有最大方差的成分",rnd_pca.components_)
print("所保留的n个成分各自的方差",rnd_pca.explained_variance_)
print("噪声方差大小",rnd_pca.noise_variance_)
```

```python
#核PCA(Kernel PCA)
# 读取数据
# data=pd.read_csv('cym_data_pca.csv',index_col='id')

#生成Swiss roll数据
from sklearn.datasets import make_swiss_roll
data=make_swiss_roll(n_samples=1000, noise=0.0, random_state=None)
X=data[0]
y=data[1]
#画3维图
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
ax = plt.subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2],c=y)
plt.show()
#kPCA
from sklearn.decomposition import KernelPCA
rbf_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=0.04)
X_reduced = rbf_pca.fit_transform(X)

from sklearn.datasets import fetch_mldata
mnist = fetch_mldata('MNIST original')
X,y = mnist["data"],mnist["target"]

from sklearn.decomposition import KernelPCA
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
clf = Pipeline([
("kpca", KernelPCA(n_components=2)),
("log_reg", LogisticRegression())
])
param_grid = [{
"kpca__gamma": np.linspace(0.03, 0.05, 10),
"kpca__kernel": ["rbf", "sigmoid"]
}]
grid_search = GridSearchCV(clf, param_grid, cv=3)
grid_search.fit(X, y)
print(grid_search.best_params_)
```
